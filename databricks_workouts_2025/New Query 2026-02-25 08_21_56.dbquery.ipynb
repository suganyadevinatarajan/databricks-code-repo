```sql



create or replace table drug_patients_wide_fact as
--This is my Star Schema Model Query
SELECT d.*,p.*,decimal(p.TotalCharges) as TotalCharges_decimal 
FROM catalog1_we4.schema1_we4.drug_info d inner join catalog1_we4.schema1_we4.patients p 
on d.uniqueID=p.drugid
where trim(TotalCharges)<>'';

create or refresh materialized view drug_patients_wide_fact_mat_vw as 
select * from drug_patients_wide_fact;
```
